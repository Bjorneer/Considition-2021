```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import random

f = pd.read_csv('visualization.txt')
size = f.iloc[0]
boxes = f.iloc[2:]
maxBox = f.iloc[1]

n_voxels = np.ones((maxBox[0], maxBox[1], maxBox[2]), dtype=bool)
colors = np.empty(n_voxels.shape, dtype=object)
colors[:,:,:] = '#ffd65d'
for idx, element in boxes.iterrows():
    n_voxels[element['x']:element['length'] + element['x'], element['y']:element['width'] + element['y'], element['z']:element['height'] + element['z']] = False

ax = plt.figure().add_subplot(projection='3d')
ax.voxels(n_voxels, facecolors=colors)

plt.show()
```
